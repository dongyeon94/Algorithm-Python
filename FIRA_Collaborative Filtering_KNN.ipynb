```python
import pandas as pd
import numpy as np
```

## 영화 - 유저 레이팅 데이터 불러오기

```python
df = pd.read_excel('movies.xlsx')
```

```python
df
```

## Sparse Matrix 형태로 변환

```python
df1 = df.groupby(['title','userId'],as_index = False)['rating'].mean().pivot('title','userId').fillna(0)['rating'].transpose()
```

```python
df1
```

```python
df1.values
```

## KNN 수행

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=30)
```

```python
neigh.fit(df1.values)
```

### 영화 별 rating의 합과 영화 별 rating의 개수를 나눈다

```python
pred_rating = np.array(df1.iloc[neigh.kneighbors([df1.values[user]])[1][0]].sum().values)/np.array(df1.iloc[neigh.kneighbors([df1.values[user]])[1][0]].astype(bool).sum(axis=0).values)
```

```python
pred_rating
```

## 추천하고자 하는 user와 비슷한 영화를 관람한 다른 user의 rating을 기반으로 예상 평점 예측

```python
import math
result_title = []
result_rating = []
for i in range(len(pred_rating)):
    if(math.isnan(pred_rating[i]) != True):
        result_title.append(df1.iloc[neigh.kneighbors([df1.values[user]])[1][0]].astype(bool).sum(axis=0).index[i])
        result_rating.append(pred_rating[i])
```

```python
df_result = pd.DataFrame({'title':result_title,'rating':result_rating}).sort_values(by = 'rating', ascending=False)
```
