Iris dataset을 K-means 알고리즘을 이용하여 클러스터링

```python
list(df)
```

```python
df[['sepal length', 'sepal width', 'petal length', 'petal width']]
```

```python
kmeans = KMeans(n_clusters=3,algorithm='auto')
kmeans.fit(df[['sepal length', 'sepal width', 'petal length', 'petal width']])
predict = pd.DataFrame(kmeans.predict(df[['sepal length', 'sepal width', 'petal length', 'petal width']]))

predict_name = []
for i in predict[0]:
    predict_name.append(str(i)+'-')
predict_name = pd.DataFrame(predict_name)
    
predict_name.columns=['predict']
predict = pd.concat([df[['sepal length', 'sepal width', 'petal length', 'petal width']],predict_name],axis=1)
```

Crosstab을 이용하여 실제 label과 clustering 결과 비교

(주의! Clustering은 unsupervised 방식임으로 보통의 경우 label이 존재하지 않는다)

```python
pd.crosstab(df['class'],predict['predict'])
```

클러스터링 결과와 4개의 feature들의 조합을 scatter matrix로 plot

```python
sns.set(style="ticks")

sns.pairplot(predict, hue="predict", diag_kind="kde")
```

```python
centers = pd.DataFrame(kmeans.cluster_centers_,columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
```

클러스터링 결과를 2개의 선택된 feature(sepal length, sepal width)를 이용하여 plot

```python
a = predict[predict['predict'] =='0-']
b = predict[predict['predict'] =='1-']
c = predict[predict['predict'] =='2-']

a_sc = plt.scatter(a['sepal length'],a['sepal width'], marker='o', color='b')
b_sc = plt.scatter(b['sepal length'],b['sepal width'], marker='x', color='g')
c_sc = plt.scatter(c['sepal length'],c['sepal width'], marker='v', color='k')
plt.scatter(centers['sepal length'],centers['sepal width'],marker='D',c='r')

plt.grid(True)
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Iris dataset", fontsize= 20)
```

클러스터링 결과를 2개의 선택된 feature(petal length, petal width)를 이용하여 plot

```python
a = predict[predict['predict'] =='0-']
b = predict[predict['predict'] =='1-']
c = predict[predict['predict'] =='2-']

a_sc = plt.scatter(a['petal length'],a['petal width'], marker='o', color='b')
b_sc = plt.scatter(b['petal length'],b['petal width'], marker='x', color='g')
c_sc = plt.scatter(c['petal length'],c['petal width'], marker='v', color='k')
plt.scatter(centers['petal length'],centers['petal width'],marker='D',c='r')

plt.grid(True)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("Iris dataset", fontsize= 20)
```

Cluster inertia를 이용하여 적정 k 갯수 파악

```python
ks = range(1,10)
inertias = []

for k in ks:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df[['sepal length', 'sepal width', 'petal length', 'petal width']])
    inertias.append(kmeans.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
